```python
#question-3

#subtask-1
class Dates:
    def __init__(self,date):
        self.date = date
    
    #subtask-2
    @classmethod
    def toDashDate(cls,slash):
        dash = ""
        for i in slash:
            if i != "/":
                dash+=i
            else:
                dash+="-"
        return Dates(dash)

    #subtask-3
    def getDate(self):
        return self.date

date1 = Dates("05-09-2020")
dateFromDB = "05/09/2020"
date2= Dates.toDashDate(dateFromDB)
if(date1.getDate() == date2.getDate() ):
    print("Equal")
else:
    print("Unequal")
    
#subtask-4
print("""\n\nExplanation: The reason why in a conditional statement it prints 'Equal' because both date1.name and date2.name
object has the same attributes and same string value. And the reason why they have same attributes and same string value because 
date2 object parameter has been changed from slash to dash.That is why when both the object's get.Date() method is being compared, 
'Equal' is printed as output.""")
```

```python
#question-4

from math import pi
#subtask-1
class Circle:
    #subtask-2
    def __init__(self,radius):
        self.__radius = radius
    
    #subtask-3
    def getRadius(self):
        return self.__radius

    #subtask-4
    def area(self):
        a=str(pi * self.__radius**2)
        return a
    
    #subtask-5
    def __add__(self,other):
        return Circle(self.__radius+other.__radius)

c1 = Circle(4)
print("First circle radius:" , c1.getRadius())
print("First circle area:" ,c1.area())
c2 = Circle(5)
print("Second circle radius:" ,c2.getRadius())
print("Second circle area:" ,c2.area())
c3 = c1 + c2
print("Third circle radius:" ,c3.getRadius())
print("Third circle area:" ,c3.area())
```

```python
#question-5
#subtask-1
class Dolls:
    #subtask-2
    def __init__(self,name,price,dollamount=1):
        self.name = name
        self.price = price
        self.dollamount = dollamount

    #subtask-3
    def detail(self):
        if self.dollamount == 1:
            return("Doll: "+self.name+"\nTotal Price: "+str(self.price)+" taka")
        else:
            return("Dolls: "+self.name+"\nTotal Price: "+str(self.price)+" taka")

    #subtask-4
    def __gt__(self,other):
        if self.price > other.price:
            return True
        else:
            return False
    
    #subtask-5
    def __add__(self,other):
        a = Dolls(self.name+" "+other.name,self.price+other.price,2)
        return a

obj_1 = Dolls("Tweety", 2500)
print(obj_1.detail())
if obj_1 > obj_1:
    print("Congratulations! You get the Tweety as a gift!")
else:
    print("Thank you!")
print("=========================")
obj_2 = Dolls("Daffy Duck", 1800)
print(obj_2.detail())
if obj_2 > obj_1:
    print("Congratulations! You get the Tweety as a gift!")
else:
    print("Thank you!")
print("=========================")
obj_3 = Dolls("Bugs Bunny", 3000)
print(obj_3.detail())
if obj_3 > obj_1:
    print("Congratulations! You get the Tweety as a gift!")
else:
    print("Thank you!")
print("=========================")
obj_4 = Dolls("Porky Pig", 1500)
print(obj_4.detail())
if obj_4 > obj_1:
    print("Congratulations! You get the Tweety as a gift!")
else:
    print("Thank you!")
print("=========================")
obj_5 = obj_2 + obj_3+obj_4
print(obj_5.detail())
if obj_5 > obj_1:
    print("Congratulations! You get the Tweety as a gift!")
else:
    print("Thank you!")

    
```
